Goal: Train an Random Forest regressor on Position_Salaries.csv, make the single prediction for level 6.5, and produce a high-resolution plot

##Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

##Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
```

```python
print(dataset)
```

```python
dataset.head()
```

##Creating the matrix of features

```python
X = dataset.iloc[:, 1:-1].values
```

```python
print(X)
```

```python
X.shape
```

##Creating the dependent variable vector

```python
y = dataset.iloc[:, -1].values
```

```python
print(y)
```

```python
y.shape
```

##Building the model

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
```

##Training the model

```python
regressor.fit(X, y)
```

##Making a prediction

```python
print(regressor.predict([[6.5]]))
```

##Visualizing the results in higher resolution

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Position Levels vs Salaries')
plt.xlabel('Position Levels')
plt.ylabel('Salaries')
plt.show()
```
