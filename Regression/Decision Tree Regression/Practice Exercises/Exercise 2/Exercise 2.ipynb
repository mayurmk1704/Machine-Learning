#Exercise: Property Age vs House Value

Goal: Build a decision tree regression model to predict house values based on property age, capturing the non-linear depreciation and appreciation patterns.

##Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

##Importing the dataset

```python
dataset = pd.read_csv('property_value_patterns.csv')
```

```python
print(dataset)
```

##Creating the matrix of features

```python
X = dataset.iloc[:, :-1].values
```

```python
print(X)
```

##Creating the dependent variable vector

```python
y = dataset.iloc[:, -1].values
```

```python
print(y)
```

##Building the model

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
#max_depth = 2
regressor_2 = DecisionTreeRegressor(max_depth=2, random_state=0)
```

```python
#max_depth = 3
regressor_3 = DecisionTreeRegressor(max_depth=3, random_state=0)
```

```python
#max_depth = 5
regressor_5 = DecisionTreeRegressor(max_depth=5, random_state=0)
```

##Training the model

```python
#max_depth = 2
regressor_2.fit(X, y)
```

```python
#max_depth = 3
regressor_3.fit(X, y)
```

```python
#max_depth = 5
regressor_5.fit(X, y)
```

##Making a prediction

```python
#Predict value for 7-year-old property
print(regressor_2.predict([[7]]))
#Predict value for 45-year-old property
print(regressor_2.predict([[45]]))
```

```python
#Predict value for 7-year-old property
print(regressor_3.predict([[7]]))
#Predict value for 45-year-old property
print(regressor_3.predict([[45]]))
```

```python
#Predict value for 7-year-old property
print(regressor_5.predict([[7]]))
#Predict value for 45-year-old property
print(regressor_5.predict([[45]]))
```

##Visualizing the results in higher resolution

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor_2.predict(X_grid), color='blue')
plt.plot(X_grid, regressor_3.predict(X_grid), color='black')
plt.plot(X_grid, regressor_5.predict(X_grid), color='green')
plt.title('Property Age vs House Value')
plt.xlabel('Property Age')
plt.ylabel('House value')
plt.show()
```
